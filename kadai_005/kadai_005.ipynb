```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,3,tight_layout=True)

y0 = [4, 6, -3, 1]
axes[0,0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 = (np.sin(x1)/2) + 1

axes[0,1].plot(x1, y1, linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(0, 3), yticks=np.arange(0, 3))

x2 = np.random .normal(30, 10, 100)
plt.xlim(0, 60)
axes[0,2].hist(x2, bins=30)

np.random.seed(10)
x = 4 + np.random.normal(0, 3, 20)
y = 4 + np.random.normal(0, 3, len(x))
axes[1,0].scatter(x, y)

plt.show()

```
